```python
import pandas as pd

data = {
    'Wind_Speed': ['Fast', 'Slow', 'Fast', 'Medium', 'Medium', 'Slow', 'Fast', 'Medium', 'Slow', 'Slow', 'Slow', 'Fast', 'Fast', 'Medium', 'Slow'],
    'Air_Type': ['Moist', 'Dry', 'Dry', 'Humid', 'Moist', 'Humid', 'Moist', 'Moist', 'Dry', 'Dry', 'Humid', 'Dry', 'Humid', 'Humid', 'Moist'],
    'Cloud_Type': ['Cumulonimbus', 'Clear', 'Broken', 'Broken', 'Cumulonimbus', 'Cumulonimbus', 'Clear', 'Broken', 'Clear', 'Clear', 'Broken', 'Clear', 'Broken', 'Cumulonimbus', 'Cumulonimbus'],
    'Rain': ['Yes', 'No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'No', 'No', 'No', 'No', 'Yes', 'Yes', 'Yes']
}

dataframe = pd.DataFrame(data)
print(len(dataframe))
display(dataframe)

#Calculate priror probabilities
p_yes = len(dataframe[dataframe['Rain'] == 'Yes'])/len(dataframe)
print("P(Yes) = ", p_yes)
p_no = len(dataframe[dataframe['Rain'] == 'No'])/len(dataframe)
print("P(No) = ", p_no)

#Calculate the likelihood probabilities
p_Fast_Yes = (len(dataframe[(dataframe['Wind_Speed'] == 'Fast') & (dataframe['Rain'] == 'Yes')])+1) / (len(dataframe[dataframe['Rain'] == 'Yes'])+1)
p_Fast_No = (len(dataframe[(dataframe['Wind_Speed'] == 'Fast') & (dataframe['Rain'] == 'No')])+1) / (len(dataframe[dataframe['Rain'] == 'No'])+1)
print("P(Fast|Yes) = ", p_Fast_Yes)
print("P(Fast|No) = ", p_Fast_No)
p_Dry_Yes = (len(dataframe[(dataframe['Air_Type'] == 'Dry') & (dataframe['Rain'] == 'Yes')])+1) / (len(dataframe[dataframe['Rain'] == 'Yes'])+1)
p_Dry_No = (len(dataframe[(dataframe['Air_Type'] == 'Dry') & (dataframe['Rain'] == 'No')])+1) / (len(dataframe[dataframe['Rain'] == 'No'])+1)
print("P(Dry|Yes) = ", p_Dry_Yes)
print("P(Dry|No) = ", p_Dry_No)
p_Broken_Yes = (len(dataframe[(dataframe['Cloud_Type'] == 'Broken') & (dataframe['Rain'] == 'Yes')])+1) / (len(dataframe[dataframe['Rain'] == 'Yes'])+1)
p_Broken_No = (len(dataframe[(dataframe['Cloud_Type'] == 'Broken') & (dataframe['Rain'] == 'No')])+1) / (len(dataframe[dataframe['Rain'] == 'No'])+1)
print("P(Broken|Yes) = ", p_Broken_Yes)
print("P(Broken|No) = ", p_Broken_No)

#Calculate the posterior probabilities
posterior_yes = p_yes*p_Fast_Yes*p_Dry_Yes*p_Broken_Yes
print(posterior_yes)
posterior_no = p_no*p_Fast_No*p_Dry_No*p_Broken_No
print(posterior_no)

#Total probability
total_probability = posterior_yes + posterior_no
print(total_probability)

rain_probability = posterior_yes/total_probability
print("Probability of Rain = ", rain_probability)

if rain_probability > 0.5:
    print("It will rain")
else:
    print("It will not rain")
```
