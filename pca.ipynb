```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Loading the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# The shape of data set before applying pca is 
print("The shape of the dataset before applying pca is "+ str(X_scaled.shape))

# Visualizing the original data in 2D (first two features)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for i, target_name in enumerate(target_names):
    plt.scatter(X_scaled[y == i, 0], X_scaled[y == i, 1], label=target_name, alpha=0.6)
plt.title('Original Data (First Two Features)')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()

# Applying PCA to reduce the dimensionality to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# To show data after applying pca 
print("The data shape after applying pca is :" + str(X_pca.shape))

# Visualizing the data after applying PCA
plt.subplot(1, 2, 2)
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name, alpha=0.6)
plt.title('Data After PCA (2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

# Showing the plots
plt.suptitle('Iris Dataset: Before and After PCA')
plt.tight_layout()
plt.show()
```
