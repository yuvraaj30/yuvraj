```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.datasets import make_blobs

# File path (commented out)
# file_path = r"C:\Users\sarth\Downloads\Mall_Customers.csv"  # Enter the path of the file


# Use an in-built dataset for clustering (make_blobs)
X, _ = make_blobs(n_samples=500, centers=6, random_state=42)

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply KMeans
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans.fit(X_scaled)

# Evaluate model with silhouette score
score = silhouette_score(X_scaled, kmeans.labels_)
print("Silhouette Score:", score)

# Plot clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='viridis', marker='o', edgecolor='k', s=50, alpha=0.6)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title('K-Means Clustering: In-built Dataset (make_blobs)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```
